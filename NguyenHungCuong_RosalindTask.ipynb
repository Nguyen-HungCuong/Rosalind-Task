#### Question 10: GC: Computing GC *Content*

```python
from Bio import SeqIO

def GCcontent(seq):
    return(seq.count('G') + seq.count('C')) * 100/len(seq)

path = '/content/rosalind_gc.fasta'
dict_seq = {}
for seq_record in SeqIO.parse(path, "fasta"):
  header = seq_record.id
  seq = str(seq_record.seq)
  gc_content = GCcontent(seq)
  dict_seq[header] = gc_content

highest_gccontent = max(dict_seq,key=dict_seq.get)

print (highest_gccontent + "\n")
print (dict_seq[highest_gccontent])
```

#### Question 19: REVP: Locating Restriction Sites

```python
def complement(dna):
  convert     = dna.maketrans('ATGC','TACG')
  dna_comp    = dna.translate(convert)
  return dna_comp

def LRS(forward_strand):
  result = []
  comp_strand = complement(forward_strand)
  LRS_length = [length for length in range(4,13)]
  for i in LRS_length:
    for j in range(len(forward_strand)):
      if (forward_strand[j:j+i] == comp_strand[j:j+i][::-1] and len(forward_strand[j:j+i]) == i):
        result.append(str(j+1)+" "+str(i))
  return result
```

#### Question 34: BA1B:Find the Most Frequent Words in a String

```python
def frequency_max(seq,k):
  # Creating list of k mer
  kmer_list = []
  for i in range(len(seq)):
    if len(seq[i:i+k]) == k:
      kmer_list.append(seq[i:i+k])
  # Couting k mer and store into a dictionary
  k_mer_dict={}
  for k_mer in kmer_list:
    if k_mer not in k_mer_dict:
      k_mer_dict[k_mer] = 1
    else:
      k_mer_dict[k_mer] +=1
  # Finding the maximum k mer
  k_mer_max = max(k_mer_dict,key=k_mer_dict.get)
  # Returning all the maximum k mer
  result = []
  for k_mer, count in k_mer_dict.items():
    if k_mer_dict[k_mer] == k_mer_dict[k_mer_max]:
      result.append(k_mer)
  return result
```

```python
with open('/content/rosalind_ba1b.txt','r') as f:
  contents  = f.readlines()
  sequence  = contents[0]
  k         = str(contents[1])

results = frequency_max(sequence,14)
for answer in results:
  print(answer)
```

```python
def fibonacci(n):
  if (n < 0):
      return -1
  elif (n == 0 or n == 1):
      return n
  else:
      return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(21)
```
